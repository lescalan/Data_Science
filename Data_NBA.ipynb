```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/final_data.csv")

print(df.info())
print(df.head())
print(df.describe())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['TEAM_ID'], bins=30, kde=True)
plt.title('Histograma de puntos por equipos')
plt.xlabel('TEAM_ID')
plt.ylabel('Puntos')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='AST', y='FGM', data=df)
plt.title('Relaci√≥n entre asistencias y puntos')
plt.xlabel('Asistencias')
plt.ylabel('Puntos')
plt.show()
```
